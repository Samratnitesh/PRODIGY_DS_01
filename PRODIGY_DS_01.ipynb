```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## About the Dataset
The current US Census Bureau world population estimate in June 2019 shows that the current global population is 7,577,130,400 people on earth, which far exceeds the world population of 7.2 billion in 2015. Our own estimate based on UN data shows the world's population surpassing 7.7 billion.

China is the most populous country in the world with a population exceeding 1.4 billion. It is one of just two countries with a population of more than 1 billion, with India being the second. As of 2018, India has a population of over 1.355 billion people, and its population growth is expected to continue through at least 2050. By the year 2030, the country of India is expected to become the most populous country in the world. This is because India’s population will grow, while China is projected to see a loss in population.

In this Dataset, we have Historical Population data for every Country/Territory in the world by different parameters like Area Size of the Country/Territory, Name of the Continent, Name of the Capital, Density, Population Growth Rate, Ranking based on Population, World Population Percentage, etc.

Dataset Glossary (Column-Wise)
<br>
**Rank:**Rank by Population.

### Distribution of countries across continents

```python
# Plot of countries across continents
grouped_df_continent = df.groupby('Continent')['Country/Territory'].apply(list).reset_index()
```

```python
sns.barplot(x=grouped_df_continent["Continent"], y=grouped_df_continent["Country/Territory"].str.len(), edgecolor='black')
plt.title("Countries across Continents")

plt.show()
```

### Countries and Population

```python
df_by_population = df.sort_values(by='Rank')
df_by_population.head(10)
```

```python
# Plot for Population Distribution by Continent
population = ["2022 Population","2020 Population","2015 Population","2010 Population","2000 Population","1990 Population","1980 Population","1970 Population"]
colors = ["olive", "darkblue", "orange", "green", "skyblue", "purple", "grey", "maroon"]
grouped_df = df.groupby('Continent').sum(numeric_only=True)
l=0
for i in population:
    grouped_df.sort_values(by=i, ascending=False)
    plt.figure(figsize=(7, 6))
    plt.bar(grouped_df.index, grouped_df[i], color=colors[l], edgecolor='black')
    plt.xlabel('Continent')
    plt.ylabel(i)
    plt.title(f'Population Distribution by Continent {i}')
    plt.xticks(rotation=45)
    l = l+1
    for j, v in enumerate(grouped_df[i]):
        plt.text(j, v + 500000, str(v), ha='center', va='bottom', fontsize=9)

plt.show()
```

```python
ax = sns.barplot(x=df_by_population["Country/Territory"][:10], y=df_by_population["World Population Percentage"], data=df_by_population, edgecolor='black')
plt.xticks(rotation=45)
plt.title("Bar Plot of top Countries with most population")
plt.show()
```

```python
# Plot of population distribution by top ten countries
population = ["2022 Population","2020 Population","2015 Population","2010 Population","2000 Population","1990 Population","1980 Population","1970 Population"]
colors = ["peru", "gold", "salmon", "aquamarine", "coral", "violet", "rosybrown", "maroon"]
j=0
for i in population:
#     grouped_df.sort_values(by=i, ascending=False)
    plt.figure(figsize=(7, 6))
    plt.bar(df_by_population["Country/Territory"][:10], df_by_population[i][:10], color=colors[j], edgecolor='black')
    plt.xlabel('Countries')
    plt.ylabel(i)
    plt.title(f'Population Distribution by Countries {i}')
    plt.xticks(rotation=45)
    j = j+1

plt.show()
```

```python
# Population growth rate of coutries

plt.figure(figsize=(20, 8))
plt.plot(df_by_population['Country/Territory'][:20], df_by_population['Growth Rate'][:20], marker='o', markersize=8, color='skyblue', linewidth=2)
plt.xlabel('Country/Territory')
plt.ylabel('Growth Rate')
plt.title('Population Growth Rate for 20 Countries')
plt.xticks(rotation=90)
plt.grid(True)

plt.show()
```

```python
# Population across the continents from year 1970 to 2022
data = {
    "Continent": df['Continent'],
    "1970 Population": df['1970 Population'],
    "1980 Population": df['1980 Population'],
    "1990 Population": df['1990 Population'],
    "2000 Population": df['2000 Population'],
    "2010 Population": df['2010 Population'],
    "2015 Population": df['2015 Population'],
    "2020 Population": df['2020 Population'],
    "2022 Population": df['2022 Population']
}
df_pop_by_continent = pd.DataFrame(data)

years = df_pop_by_continent.columns[1:]  
plt.figure(figsize=(20, 8))
for continent in df_pop_by_continent["Continent"].unique():
    continent_data = df_pop_by_continent[df_pop_by_continent["Continent"] == continent].iloc[0, 1:]
    plt.plot(years, continent_data, label=continent)

plt.title('Population by Continent (1970-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```

### Distribution of Countries with Area

```python
df_by_areas = df.sort_values(by='Area (km²)', ascending=False)
df_by_areas.head(10)
```

```python
flights_wide = pd.DataFrame(df_by_areas)
sns.barplot(x=df_by_areas["Country/Territory"][:10], y=df_by_areas["Area (km²)"][:10], data=flights_wide, edgecolor='black')
plt.xticks(rotation=45)
plt.title("Bar plot of Total Land Area and Countries")
plt.show()
```

```python
df_most_density = df.sort_values(by="Density (per km²)", ascending=False)
df_most_density.head(10)
```

```python
# Plotting the countries with the most density per km^2
flights_wide = pd.DataFrame(df_most_density)
plt.figure(figsize=(9, 6))
sns.barplot(x=df_most_density["Density (per km²)"][:20], y=df_most_density["Country/Territory"][:20], edgecolor='black')
plt.title("Bar Plot of Population Density and Countries")

plt.show()
```
